```python
import calendar
from datetime import datetime


def getDates(month,year):
    num_days = calendar.monthrange(year, month)[1]
    dates = [datetime(year, month, day).strftime('%Y-%m-%d') for day in range(1, num_days + 1)]
    return dates

getDates(2,2024)
```

```python
import os
def getFolders(path):
    items = []
    with os.scandir(path) as entries:
        for entry in entries:
            if entry.is_dir():
                item_dict = {
                    'name': entry.name,
                    'path': entry.path,
                    'is_file': entry.is_file(),
                    'is_dir': entry.is_dir()
                }
                items.append(item_dict)
    return items

def getFiles(path):
    items = []
    with os.scandir(path) as entries:
        for entry in entries:
            if entry.is_file():
                _, ext = os.path.splitext(entry.name)
                item_dict = {
                    'name': entry.name,
                    'path': entry.path,
                    'type': (ext[1:]).strip() # File extension without dot
                }
                items.append(item_dict)
    return items
getFolders('D:\\DS_Data Driven Stock Analysis\\stock_data')
```

```python
import yaml
import pandas as pd

def getDataFrame(path):
    data_files=getFiles(path)
    print(len(data_files))
    master_df=pd.DataFrame()
    for i in data_files:
        if i['type']=='yaml':
            with open(i['path'],'r') as file:
                data=yaml.safe_load(file)
            df=pd.DataFrame(data)
            if not master_df.empty:
                master_df=pd.concat([master_df,df],axis=0).reset_index(drop=True)
            else:
                master_df=df.copy(deep=True)
    master_df.shape
    return master_df
getDataFrame(r'D:\DS_Data Driven Stock Analysis\stock_data\2023-10')
```
